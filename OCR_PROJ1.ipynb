```
from googletrans import Translator
import easyocr
from gtts import gTTS
from IPython.display import Audio 
```

```
reader = easyocr.Reader(['mr'])
#translator = Translator()
```

```
translator = Translator()
```

```
import PIL
from PIL import ImageDraw
im = PIL.Image.open("main-qimg-1a7a6456deed6b04e5577765224dfd4f")
im
```

```
bounds = reader.readtext("main-qimg-1a7a6456deed6b04e5577765224dfd4f", add_margin = 0.55, width_ths = 0.7, link_threshold = 0.8, decoder = 'beamsearch', )
bounds
```

```
def draw_boxes(image,bounds,color='yellow',width=2):
    draw = ImageDraw.Draw(image)
    for bound in bounds:
      p0,p1,p2,p3 = bound[0]
      draw.line([*p0,*p1,*p2,*p3,*p0], fill=color, width=width)
    return image 

draw_boxes(im, bounds)
```

```
#text_list = reader.readtext("main-qimg-1a7a6456deed6b04e5577765224dfd4f", add_margin = 0.55, width_ths = 0.7, link_threshold = 0.8, decoder = 'beamsearch', )
#text_list

text_list = reader.readtext('main-qimg-1a7a6456deed6b04e5577765224dfd4f', add_margin=0.55, width_ths=0.7, link_threshold=0.8, decoder='beamsearch',
                            blocklist='=-', detail=0)
text_list

```

```
text_comb = ' '.join(text_list)
text_comb
```

```
print(translator.detect(text_comb))
```

```
test_en = translator.translate(text_comb, src='mr')
print(test_en.text)
```

```
#English word US accent.
mr_tts = gTTS(test_en.text)
mr_tts.save('trans.mp3') 
Audio('trans.mp3', autoplay= True)
```

```
## Hindi words, hindi accent 
test_hiad = gTTS(test_hi.text, lang = 'hi')
test_hiad.save("trans4.mp3")
Audio('trans4.mp3', autoplay = True)
```
